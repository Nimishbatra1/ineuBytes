```python
X=data.iloc[:,:-1]
X
```

```python
Y=data.iloc[:,-1]
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,random_state=0,test_size=0.3)
```

```python
X_train.shape
```

```python
X_test.shape
```

## MODEL IMPLEMENTATION

### RIDGE REGRESSOR

```python
from sklearn.linear_model import Ridge
model=Ridge(alpha=0.5)
model.fit(X_train,y_train)
```

```python
y_test_predict_ridge=model.predict(X_test)
```

```python
y_train_predict_ridge=model.predict(X_train)
```

## RANDOM FOREST REGRESSOR

```python
from sklearn.ensemble import RandomForestRegressor
model2=RandomForestRegressor(n_estimators=350,max_depth=8,random_state=50)
model2.fit(X_train,y_train)
```

```python
y_test_predict=model2.predict(X_test)
```

```python
y_train_predict=model2.predict(X_train)
```

## MODEL EVALUATION

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from math import sqrt
```

### RIDGE REGRESSOR

```python
print("           Training Data Evaluation         " )
print("Mean Absolute Error : ",mean_absolute_error(y_train_predict_ridge,y_train))
print("Mean Squared Error : ",mean_squared_error(y_train_predict_ridge,y_train))
print("Root Mean Squared Error : ",sqrt(mean_squared_error(y_train_predict_ridge,y_train)))
print("R2 Score : ",r2_score(y_train_predict_ridge,y_train))
print("\n")
print("           Testing Data Evaluation         " )
print("Mean Absolute Error : ",mean_absolute_error(y_test_predict_ridge,y_test))
print("Mean Squared Error : ",mean_squared_error(y_test_predict_ridge,y_test))
print("Root Mean Squared Error : ",sqrt(mean_squared_error(y_test_predict_ridge,y_test)))
print("R2 Score : ",r2_score(y_test,y_test_predict_ridge))
```

### RANDOM FOREST REGRESSOR

```python
print("           Training Data Evaluation         " )
print("Mean Absolute Error : ",mean_absolute_error(y_train_predict,y_train))
print("Mean Squared Error : ",mean_squared_error(y_train_predict,y_train))
print("Root Mean Squared Error : ",sqrt(mean_squared_error(y_train_predict,y_train)))
print("R2 Score : ",r2_score(y_train_predict,y_train))
print("\n")
print("           Testing Data Evaluation         " )
print("Mean Absolute Error : ",mean_absolute_error(y_test_predict,y_test))
print("Mean Squared Error : ",mean_squared_error(y_test_predict,y_test))
print("Root Mean Squared Error : ",sqrt(mean_squared_error(y_test_predict,y_test)))
print("R2 Score : ",r2_score(y_test,y_test_predict))
```

```python
Models=['Ridge Regressor','Random Forest Regressor']
MAE=[mean_absolute_error(y_train_predict_ridge,y_train),
     mean_absolute_error(y_train_predict,y_train)]

MSE=[mean_squared_error(y_train_predict_ridge,y_train),
     mean_squared_error(y_train_predict,y_train)]

RMSE=[sqrt(mean_squared_error(y_train_predict_ridge,y_train)),
     sqrt(mean_squared_error(y_train_predict,y_train))]

R2_score=[r2_score(y_train_predict_ridge,y_train),
     mean_squared_error(y_train_predict,y_train)]

models_list=list(zip(Models,MAE,MSE,RMSE,R2_score))

print("Training Dataset Outcomes")
models_training=pd.DataFrame(models_list,columns=['Models','MAE','MSE','RMSE','R2_score'])
models_training
```
